```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
from time import sleep

# URL ì„¤ì • (í˜ì´ì§€ ë²ˆí˜¸ í¬í•¨)
base_url = "https://you.ctrip.com/sight/jeju1446512/134029217.html"

# HTTP ìš”ì²­ í—¤ë” ì„¤ì • (í•„ìš” ì‹œ)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# ì‚¬ìš©ì ì½”ë©˜íŠ¸ ë°ì´í„° ì¶”ì¶œ í•¨ìˆ˜
def get_comments(page):
    url = f"{base_url}?page={page}"
    response = requests.get(url, headers=headers)
    response.encoding = 'utf-8'  # ì¸ì½”ë”© ì„¤ì •
    html_content = response.text

    # BeautifulSoupìœ¼ë¡œ HTML íŒŒì‹±
    soup = BeautifulSoup(html_content, "html.parser")

    # ì½”ë©˜íŠ¸ ëª¨ë“ˆ ì„ íƒ
    div_comment_module = soup.find("div", id="commentModule")

    # ì‚¬ìš©ì ì½”ë©˜íŠ¸ ë°ì´í„° ì¶”ì¶œ
    comments = []
    if div_comment_module:
        comment_items = div_comment_module.find_all("div", class_="commentItem")
        for item in comment_items:
            user_name = item.find("div", class_="userName")
            score_info = item.find("span", class_="averageScore")
            comment_detail = item.find("div", class_="commentDetail")
            comment_time = item.find("div", class_="commentTime")

            user_name = user_name.get_text(strip=True) if user_name else "ìµëª…"
            score = score_info.get_text(strip=True) if score_info else "N/A"
            detail = comment_detail.get_text(strip=True) if comment_detail else ""
            time = comment_time.get_text(strip=True) if comment_time else ""

            comments.append({
                "user_name": user_name,
                "score": score,
                "detail": detail,
                "time": time
            })
    return comments

# ì „ì²´ ì½”ë©˜íŠ¸ ìˆ˜ì§‘
total_comments = []
for page in range(1, 11):  # 1~10í˜ì´ì§€ í¬ë¡¤ë§
    print(f"í¬ë¡¤ë§ ì¤‘: {page}í˜ì´ì§€")
    total_comments.extend(get_comments(page))
    sleep(2)  # ìš”ì²­ ê°„ ëŒ€ê¸° (ë°˜ë³µì ì¸ ìš”ì²­ìœ¼ë¡œ ì¸í•œ ì°¨ë‹¨ ë°©ì§€)

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(total_comments)

# ê²°ê³¼ ì¶œë ¥ (ì„ íƒ ì‚¬í•­)
print(df)

# CSV íŒŒì¼ë¡œ ì €ì¥
df.to_csv("comments.csv", index=False, encoding="utf-8")
```
