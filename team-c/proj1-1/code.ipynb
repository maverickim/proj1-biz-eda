```python
import matplotlib.pyplot as plt
import pandas as pd

#  
data = {
    '': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    '': [2000000, 2200000, 2500000, 2400000, 2300000, 2100000, 500000, 600000, 1700000, 1800000],
    '': [500000, 480000, 470000, 460000, 450000, 440000, 100000, 120000, 430000, 450000],
    '': [100000, 110000, 120000, 130000, 140000, 150000, 50000, 60000, 110000, 120000],
    '': [200000, 210000, 220000, 230000, 240000, 250000, 70000, 80000, 240000, 250000]
}

df = pd.DataFrame(data)
df[''] = df[['', '', '', '']].sum(axis=1)

# 
plt.figure(figsize=(12, 6))
for country in ['', '', '', '']:
    plt.plot(df[''], df[country], marker='o', label=country)

plt.plot(df[''], df[''], marker='o', linestyle='--', color='black', label='')

plt.title('     ')
plt.xlabel('')
plt.ylabel(' ')
plt.legend()
plt.grid(True)
plt.xticks(df[''])
plt.tight_layout()
plt.show()
```
