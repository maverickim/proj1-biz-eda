#Trial1 : json 형태를 개별 필드로 펼치기
- event_params, user_properties 이슈 까지 해결 해놓은 상태

```python
import pandas as pd
import json

def parse_key_value_json(json_str):
    """
    1) 문자열(json_str)이 JSON 형식이라면 로드한다.
    2) 내부에 'key', 'value'로 이뤄진 구조가 있으면 이를 {key: value} 형태로 변환한다.
    3) 그 외(단순 dict 등) 구조도 가능하면 딕셔너리 형태로 반환한다.
    4) 파싱에 실패하거나 None인 경우 빈 dict({}) 반환.
    """
    if not isinstance(json_str, str):
        return {}

    json_str = json_str.strip()
    if not json_str.startswith('{'):
        return {}

    try:
        data = json.loads(json_str)  # 문자열을 실제 JSON으로 변환
    except:
        return {}

    # 최종 결과를 담을 딕셔너리
    result = {}

    # "key-value" 구조가 배열 형태인 경우 예시 처리
    # 예: { "event_params": [ {"key": "abc", "value": {...}}, ... ] }
    if isinstance(data, dict):
        for k, v in data.items():
            # v가 list 형태이고, 안에 {"key": ..., "value": ...} 형태라면 처리
            if isinstance(v, list):
                for item in v:
                    if "key" in item and "value" in item:
                        # value 안에서 string_value, int_value 등 중 하나를 추출
                        val_dict = item["value"]
                        # 가능한 값을 하나라도 찾으면 result에 매핑
                        merged_val = (val_dict.get("string_value") or
                                      val_dict.get("int_value") or
                                      val_dict.get("float_value") or
                                      val_dict.get("double_value"))
                        result[item["key"]] = merged_val
            elif isinstance(v, dict) and "key" in v and "value" in v:
                # 단일 객체 {"key": "xxx", "value": {...}} 형태인 경우
                val_dict = v["value"]
                merged_val = (val_dict.get("string_value") or
                              val_dict.get("int_value") or
                              val_dict.get("float_value") or
                              val_dict.get("double_value"))
                result[v["key"]] = merged_val
            else:
                # 단순 딕셔너리나 일반 값이면 result[k] = v 형태로 저장
                # (필요에 따라 더 깊게 파고들 수 있음)
                result[k] = v
    return result

def expand_json_columns(df):
    """
    1) df의 모든 컬럼을 순회.
    2) 각 컬럼의 값이 '{'로 시작하는 JSON 문자열일 경우 parse_key_value_json으로 처리.
    3) 결과 딕셔너리를 펼쳐서(각 key를 새 컬럼으로) df에 추가.
    4) 기존 컬럼은 삭제.
    """
    # 컬럼 목록 사본 (반복 중 컬럼 삭제 대비)
    cols = df.columns.tolist()

    for col in cols:
        # 샘플 데이터 몇 건을 보고 JSON 구조인지 간단히 판단
        # (모든 row를 확인해도 되지만, 대체로 몇 건만 확인해도 충분)
        sample_values = df[col].dropna().head(5).astype(str)
        if any(sv.strip().startswith('{') for sv in sample_values):
            # 실제 파싱 시도
            parsed_series = df[col].apply(parse_key_value_json)
            # key-value들이 모인 시리즈를 DataFrame으로 변환(json_normalize 사용 가능)
            expanded_df = pd.json_normalize(parsed_series)

            # 컬럼명 충돌 방지를 위해 기존 col 이름을 prefix로 붙이거나 suffix를 붙인다.
            expanded_df = expanded_df.add_prefix(f"{col}.")

            # 기존 df에 병합
            df = pd.concat([df, expanded_df], axis=1)
            # 원본 컬럼은 삭제
            df.drop(columns=[col], inplace=True)

    return df

# --- 사용 예시 ---
df_flattened = expand_json_columns(df)
```

# 따라서 'items'라는 column만 별도로 파싱 진행 (output 27개)
# 다른 column은 위에 있었던 기존 코드 (output 184개 )로 파싱 진행
# -> 최종 211개 column

```python
import pandas as pd
import json

################################################
# 1) items 전용 파싱 코드
################################################
def parse_items_column(json_str):
    """
    df['items'] 내 문자열이:
    {
      "items": [{
         "item_id": "...",
         "item_name": "...",
         ...
      }]
    }
    형태라고 가정.
    1) JSON 파싱 후, "items" 키 아래 리스트가 있으면 첫 번째 아이템(dict)을 반환.
    2) 그 외(파싱 실패, 리스트 비었음 등)에서는 {} 반환.
    """
    if not isinstance(json_str, str):
        return {}

    json_str = json_str.strip()
    if not json_str.startswith('{'):
        return {}

    try:
        data = json.loads(json_str)  # JSON 문자열 → 파이썬 dict
    except:
        return {}

    items_list = data.get("items", [])
    if len(items_list) > 0 and isinstance(items_list[0], dict):
        # 첫 번째 아이템만 펼치기
        return items_list[0]
    return {}

def expand_items_column(df):
    """
    1) df['items']를 parse_items_column으로 파싱.
    2) 각 row마다 반환된 딕셔너리를 pd.json_normalize로 펼침.
    3) 컬럼명 앞에 'items.' 접두어를 붙여 구분하고, 기존 'items' 컬럼을 삭제.
    """
    if 'items' not in df.columns:
        return df  # 'items' 컬럼이 없는 경우 그대로 반환

    # 각 row 파싱
    parsed_series = df['items'].apply(parse_items_column)

    # 파싱 결과(딕셔너리) 펼치기
    expanded_df = pd.json_normalize(parsed_series)

    # 컬럼명 충돌 방지/가독성 위해 prefix 추가
    expanded_df = expanded_df.add_prefix("items.")

    # 기존 df와 병합 후, 'items' 컬럼 삭제
    df = pd.concat([df, expanded_df], axis=1)
    df.drop(columns=['items'], inplace=True)

    return df

################################################
# 2) GA4 스타일 key-value 컬럼 파싱 코드
################################################
def parse_key_value_json(json_str):
    """
    1) 문자열(json_str)이 JSON 형식이라면 로드한다.
    2) 내부에 'key', 'value'로 이뤄진 구조가 있으면 이를 {key: value} 형태로 변환한다.
    3) 그 외(단순 dict 등) 구조도 가능하면 딕셔너리 형태로 반환한다.
    4) 파싱에 실패하거나 None인 경우 빈 dict({}) 반환.
    """
    if not isinstance(json_str, str):
        return {}

    json_str = json_str.strip()
    if not json_str.startswith('{'):
        return {}

    try:
        data = json.loads(json_str)  # 문자열을 실제 JSON으로 변환
    except:
        return {}

    # 최종 결과를 담을 딕셔너리
    result = {}

    # "key-value" 구조가 배열 형태인 경우 예시 처리
    # 예: { "event_params": [ {"key": "abc", "value": {...}}, ... ] }
    if isinstance(data, dict):
        for k, v in data.items():
            # v가 list 형태이고, 안에 {"key": ..., "value": ...} 형태라면 처리
            if isinstance(v, list):
                for item in v:
                    if "key" in item and "value" in item:
                        val_dict = item["value"]
                        merged_val = (
                            val_dict.get("string_value")
                            or val_dict.get("int_value")
                            or val_dict.get("float_value")
                            or val_dict.get("double_value")
                        )
                        result[item["key"]] = merged_val
            elif isinstance(v, dict) and "key" in v and "value" in v:
                # 단일 객체 {"key": "xxx", "value": {...}} 형태인 경우
                val_dict = v["value"]
                merged_val = (
                    val_dict.get("string_value")
                    or val_dict.get("int_value")
                    or val_dict.get("float_value")
                    or val_dict.get("double_value")
                )
                result[v["key"]] = merged_val
            else:
                # 단순 딕셔너리나 일반 값이면 result[k] = v 형태로 저장
                result[k] = v

    return result

def expand_json_columns(df):
    """
    1) df의 모든 컬럼을 순회하되, 'items' 컬럼은 건너뜀(이미 별도 처리).
    2) 각 컬럼의 값이 '{'로 시작하는 JSON 문자열일 경우 parse_key_value_json으로 처리.
    3) 결과 딕셔너리를 펼쳐서 df에 추가하고, 원본 컬럼은 삭제.
    """
    cols = df.columns.tolist()

    for col in cols:
        if col == 'items':  # items 컬럼은 이미 별도 처리
            continue

        sample_values = df[col].dropna().head(5).astype(str)
        if any(sv.strip().startswith('{') for sv in sample_values):
            # 실제 파싱 시도
            parsed_series = df[col].apply(parse_key_value_json)
            expanded_df = pd.json_normalize(parsed_series)
            expanded_df = expanded_df.add_prefix(f"{col}.")

            df = pd.concat([df, expanded_df], axis=1)
            df.drop(columns=[col], inplace=True)

    return df

################################################
# 3) 모든 컬럼을 한 번에 처리하는 함수
################################################
def expand_all_json(df):
    """
    1) 'items' 컬럼을 먼저 펼친다.
    2) 남은 JSON 컬럼(예: event_params 등)을 펼친다.
    3) 최종적으로 하나의 df를 반환한다.
    """
    df = expand_items_column(df)
    df = expand_json_columns(df)
    return df

################################################
# 4) 사용 예시
################################################

df_flattened2 = expand_all_json(df)          # items -> key-value JSON 순서로 파싱
df_flattened2.info()
```

```python
import pandas as pd
from datetime import datetime, timezone, timedelta

# KST (UTC+9) 타임존 객체
KST = timezone(timedelta(hours=9))

# 변환 대상 컬럼 목록
TARGET_TIMESTAMP_COLS = [
    'event_timestamp',
    'event_previous_timestamp',
    'user_first_touch_timestamp',
    'event_params.dynamic_link_accept_time',
    'user_properties.first_open_time'
]

def parse_microseconds_to_kst(ts_micro):
    """
    단일 값(ts_micro: 마이크로초 단위) -> KST 기준 datetime 객체로 변환
    변환 불가 시 pd.NaT 반환
    """
    if pd.isnull(ts_micro):
        return pd.NaT
    try:
        # 혹시 문자열일 수 있으므로 int로 캐스팅(정밀도 보존)
        ts_int = int(ts_micro)
        # 초 단위로 만들기
        ts_sec = ts_int / 1_000_000
        # fromtimestamp(..., tz=KST) => tz-aware datetime 객체
        dt_kst = datetime.fromtimestamp(ts_sec, KST)
        return dt_kst
    except:
        return pd.NaT

def convert_defined_cols_to_kst_datetime(df):
    """
    1) TARGET_TIMESTAMP_COLS 내의 컬럼 중 df에 존재하는 컬럼을 찾는다.
    2) 각 컬럼을 마이크로초 단위 -> KST datetime으로 변환 (tz-aware).
    3) 변환 후 df[col]은 'datetime64[ns, tz]' 또는 object(dtype=Timestamp) 형태를 가질 수 있다.
    4) 시간 차 계산 등 datetime 연산이 가능해진다.
    """
    for col in TARGET_TIMESTAMP_COLS:
        if col in df.columns:
            # 숫자(또는 문자열) -> int64로 변환 시도 (정밀도 손실 방지)
            df[col] = pd.to_numeric(df[col], errors='coerce')  # float64, NaN for invalid
            df[col] = df[col].apply(parse_microseconds_to_kst)

            # df[col]이 tz-aware datetime 객체로 구성된 Series라면,
            # pandas에서 dtype='object'로 보일 수도 있고,
            # 내부적으로 tz-aware Timestamps로 유지된다 (연산은 문제없이 가능).
    return df


# ---------------- 사용 예시 ----------------
# 가정: df_flattened2를 불러온 후
df_flattened3 = convert_defined_cols_to_kst_datetime(df_flattened2)
#
# # 변환된 후, 예: df_flattened2['event_timestamp']를 보면
# # [2024-11-13 00:06:00+09:00, ...] (표기 예시) 등의 값이 들어 있을 수 있음.
#
# # 소요 시간 계산 예시:
df_flattened3['time_diff'] = df_flattened3['event_timestamp'] - df_flattened3['event_previous_timestamp']
print(df_flattened3[['event_timestamp','event_previous_timestamp','time_diff']])
```
